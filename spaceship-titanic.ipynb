```python
from sklearn.preprocessing import MinMaxScaler 

scaler = MinMaxScaler()
columns_to_scale = ["Age", "RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]

df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
df_test[columns_to_scale] = scaler.transform(df_test[columns_to_scale])
```

```python
mapping_dict = {'Earth': 0, 'Europa': 1, 'Mars': 2}
df["HomePlanet"] = df["HomePlanet"].map(mapping_dict)
df_test["HomePlanet"] = df_test["HomePlanet"].map(mapping_dict)
```

```python
mapping_dict2 ={'TRAPPIST-1e': 0, '55 Cancri e': 1, 'PSO J318.5-22': 2}
df["Destination"] = df["Destination"].map(mapping_dict2)
df_test["Destination"] = df_test["Destination"].map(mapping_dict2)
```

```python
mapping_dict3 ={False: 0, True: 1}
df["CryoSleep"] = df["CryoSleep"].map(mapping_dict3)
df_test["CryoSleep"] = df_test["CryoSleep"].map(mapping_dict3)
```

```python
df["VIP"] = df["VIP"].map(mapping_dict3)
df_test["VIP"] = df_test["VIP"].map(mapping_dict3)
```

```python
df["Transported"] = df["Transported"].map(mapping_dict3)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model2=KNeighborsClassifier(n_neighbors=5)
model2.fit(x_train,y_train)
y_pred=model2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.svm import SVC
model3=SVC(kernel='linear')
model3.fit(x_train,y_train)
y_pred=model3.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
model4=GaussianNB()
model4.fit(x_train,y_train)
y_pred=model4.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
model5=DecisionTreeClassifier(criterion='entropy')
model5.fit(x_train,y_train)
y_pred=model5.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model6= RandomForestClassifier(n_estimators=70,criterion="entropy")
model6.fit(x_train,y_train)
y_pred=model6.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
